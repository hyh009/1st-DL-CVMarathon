```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json
url = "https://www.dcard.tw/_api/forums/pet/posts?popular=true"
res = requests.get(url)
content = json.loads(res.text)

print('å…±{}ç­†è³‡æ–™'.format(len(content)))
print('æ¬„ä½æ¨™é¡Œ = {}'.format(content[0].keys()))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for c in content:
    print('æ¨™é¡Œ: {}'.format(c['title']))
    print('è²¼æ–‡æ™‚é–“: {}'.format(c['createdAt']))
    print('ç•™è¨€äººæ•¸: {}'.format(c['commentCount']))
    print('æŒ‰è®šäººæ•¸: {}'.format(c['likeCount']))
    print('======='*5)
    
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« (æœ€æ–°æ–‡ç« )çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

#ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
popular_commentcount = 0
popular_likecount = 0
new_commentcount = 0
new_likecount = 0

#ç†±é–€æ–‡ç« 
for c in content:
    
    popular_commentcount += c['commentCount']
    popular_likecount += c['likeCount']


print('ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸: {:.1f}'.format(popular_commentcount/len(content)))
print('ç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸: {:.1f}'.format(popular_likecount/len(content)))

#éç†±é–€æ–‡ç« 

#å–å¾—éç†±é–€æ–‡ç« è³‡è¨Š
url_new = "https://www.dcard.tw/_api/forums/dcard/posts?popular=false"
res2 = requests.get(url_new)
content2 = json.loads(res2.text)

for c in content2:
    
    new_commentcount += c['commentCount']
    new_likecount += c['likeCount']


print('éç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸: {:.1f}'.format(new_commentcount/len(content2)))
print('éç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸: {:.1f}'.format(new_likecount/len(content2)))

```
